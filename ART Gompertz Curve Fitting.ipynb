```python
### Gompertz Curve Fitting
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from pandas import Series,DataFrame
import os

%matplotlib inline
```

```python
data = pd.ExcelFile('ART_coverage2.xlsx')
```

```python
dframe = data.parse('Data')
```

```python
dframe.head()
```

```python
dframe2 = dframe.pivot_table(index='Country Name',columns='Year',values='Value')
```

```python
dframe2 = dframe2.dropna(axis=0,how='all')
dframe2.head()
```

```python
dframe2 = dframe2.dropna(axis=1,how='all')
dframe2.head()
```

```python
diff = dframe2.diff(axis=1)
diff.fillna(0,inplace=True)
diff.head()
```

```python
from scipy.optimize import curve_fit
import math
from math import exp
```

```python
def gompertz(a,mu,lamb,t):
    return a*np.exp(-np.exp(mu*np.exp(1)/a*(lamb-t)+1))
```

```python
### Initial guesses
mu = diff.ix['Argentina'].max()
a = 1
lamb = 0
```

```python
### Test data
t = list(np.arange(0,15))
y = dframe2.ix['Argentina'].tolist()
```

```python
### Fit curve
popt, pcov = curve_fit(gompertz, t, y, p0=list([mu,a,lamb]))
```
