```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
from pandas import Series,DataFrame
```

```python
file = "C:\\Users\Ben\Documents\Harvard Files\Research\ART Paper\Final_Output.xlsx"
a = pd.read_excel(file,sheetname="A",header=0)
lamb = pd.read_excel(file,sheetname="Lambda",header=0)
mu = pd.read_excel(file,sheetname="Mu",header=0)
```

```python
mu_plot=pd.melt(mu,id_vars='Country Name',value_vars=['mu_gomp','mu_log','avg_change'],
                var_name='Metric',value_name='Value')
```

```python
mu_plot.replace({'Metric':{'mu_gomp':'$\mu$ - Gompertz','mu_log':'$\mu$ - Logistic',
                'avg_change':'Avg. change'}},inplace=True)
```

```python
mu_plot['Metric']=pd.Categorical(mu_plot['Metric'],['$\mu$ - Gompertz','$\mu$ - Logistic','Avg. change'])
```

```python
sns.set()
ax=sns.stripplot(x='Value',y='Country Name',data=mu_plot.sort_values(by=['Metric','Value'],ascending=[True,False]),
                 hue='Metric',size=9)
fig=ax.get_figure()
ax.set_ylabel('')
ax.set_xlabel('Rate of change',fontsize=16)
fig.set_size_inches((10,12))
ax.tick_params(axis='y',labelsize=13)
ax.tick_params(axis='x',labelsize=16)
ax.legend(fontsize=14,loc=7,frameon=True,facecolor='white')
fig.savefig('C:\\Users\Ben\Documents\Harvard Files\Research\ART Paper\mu_countries.png',dpi=300,bbox_inches='tight')
plt.show()
```
